# Convert WAV to FLAC's

```python
from glob import glob
def convertWavsToFlac(array, SAMPLE_RATE=48000):
    from pydub import AudioSegment
    for path in array:
        if path.endswith('.wav'):
            song = AudioSegment.from_wav(path)
            song.export(path.replace(".wav",".flac"),format = "flac", bitrate=str(SAMPLE_RATE))
```

```python
wav_directories = """

""".split("\n")[1:-1]

for dir in wav_directories:
    print(dir)
    file_paths = glob(dir+"/**/*.wav", recursive=True)
    convertWavsToFlac(file_paths)
```

# Concatenate Wavs and output Audacity Label based on Filename

```python
from glob import glob
import librosa
import os
import numpy as np

dir = "/media/cookie/Samsung 860 QVO/Datasets_In_Progress/TF2Medic"
SAMPLE_RATE = 44100

dir_out = os.path.join(dir, 'Labels')
if not os.path.exists(dir_out):
    os.makedirs(dir_out)
file_paths = glob(dir+"/*.wav")
time_ = 0
sound = list()
string_out = ''
for file_path in file_paths:
    data = librosa.to_mono(librosa.core.load(file_path, sr=SAMPLE_RATE)[0])
    basename = os.path.basename(file_path).split('.wav')[0]
    sound.extend(data)
    string_out = f"{string_out}{time_}\t"
    time_+=len(data)/SAMPLE_RATE
    string_out = f"{string_out}{time_}\t{basename}\n"

librosa.output.write_wav(os.path.join(dir_out,"concat.wav"), np.asarray(sound), SAMPLE_RATE)
text_file = open(os.path.join(dir_out,"label.txt"), "w", encoding="utf-8")
text_file.write(string_out); text_file.close()
```

```python
from glob import glob
import librosa
import soundfile as sf
import os
import numpy as np
import datetime
directory = r"/media/cookie/Samsung 860 QVO/Datasets_In_Progress/TF2Medic/Labels"
label_name = r"label.txt"
wav_name = r"concat.wav"
output_name = r"Sliced"
voice = "Medic"
emotion = "NA"
noise = ""
SAMPLE_RATE = 44100

if not os.path.exists(os.path.join(directory,output_name)):
    os.makedirs(os.path.join(directory,output_name))
sound = librosa.to_mono(librosa.core.load(os.path.join(directory, wav_name), sr=SAMPLE_RATE)[0])
for line in label.split("\n"):
    start_time, end_time, quote = line.split("\t")
    time_stamp = str(datetime.timedelta(seconds=round(float(start_time),0))).replace(":","_")
    if len(time_stamp) < 8: time_stamp = "0"+time_stamp
    try:
        start_step = int(float(start_time)*SAMPLE_RATE)
        end_step = int(float(end_time)*SAMPLE_RATE)
        sf.write(os.path.join(directory,output_name,f"{time_stamp}_{voice}_{emotion}_{noise}_{quote}.wav"), np.asarray(sound[start_step:end_step]), SAMPLE_RATE, format='flac', subtype='PCM_24')
        text_file = open(os.path.join(directory,output_name,f"{time_stamp}_{voice}_{emotion}_{noise}_{quote}.txt"), "w", encoding="utf-8")
        text_file.write(quote); text_file.close()
    except Exception as ex:
        print(ex)
```
